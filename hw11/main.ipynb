Программа рекомендаций использует преобразованное описание фильма-примера в виде разреженной матрицы веторов для поиска похожих описаний фильмов в базе данных фильмов.

# 3 Назначение разработки

##3.1 Функциональное назначение

Данная программа позволяет преобразовать обзоры фильмов в векторную форму и выполнить поиск похожих результатов по заданному поисковому запросу.

#4 Требования к программе

##4.1 Требования к функциональным характеристикам

Программа состоит из серверной и клиентской части, между которыми налажено взаимодействие

### 4.1.1 Требования к серверной части

На серверной части должен быть реализован алгоритм определения семантической близости обзоров фильмов с заданным образцом. Для определения семантической близости используется косинусное расстояние между векторными представлениями текстов. Так же серверная часть должна поддерживать хранение базы данных в формате разреженных векторов (например Postgres).

### 4.1.2. Требование к взаимодействию клиентской и серверной частей

Взаимодействие между клиентской и серверной частями должно осуществляться
посредством HTTP-запросов. При получении GET-запроса от клиента, сервер должен
ответить сообщением в формате JSON, содержащим список фильмов с названием, описанием, рейтингом imdb и списком 5 снимавшихся актеров.

### 4.1.3. Требования к клиентской части

Клиентская часть должна быть реализована в виде веб-приложения, запускаемого в
браузере, и представлена в виде интерактивной таблицы с сортированным по убыванию рейтинга imdb списком фильмов в предложении. При нажатии на название, описание или постер фильма должен совершаться переход на страницу с этим фильмом, расположенным на интернет ресурсе "тестмуви.ру"

Веб-приложение должно предоставлять следующие возможности:
- ввод названия фильма-образца для поиска похожих;
- предложение в виде выпадающего списка выбрать фильм-образец из существующей базы по введеному названию;
- ввод минимального рейтинга для отбора фильмов;
- иметь возможность ввести фильм-образец повторно на той же странице после отображения списка по предыдущему запросу.
