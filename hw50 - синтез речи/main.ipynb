# Установка языковых пакетов для перевода

Перевод будет осуществляться через модель `argostranslate`, запуск которой осуществляется локально. Переводы для пар без английского осуществляются путем перевода оригинала на английский и с английского на целевой.

````python
%%writefile install_argos.py
import argostranslate.package
import argostranslate.translate
import sys

# Языковые пары для установки
# Формат: ("from", "to")
LANG_PAIRS = [
    ("de", "en"),
    ("fr", "en"),
    ("ru", "en"),
    ("en", "de"),
    ("en", "fr"),
    ("en", "ru"),
]

def main():
    print("🔎 Fetching available ArgosTranslate packages...")
    pkgs = argostranslate.package.get_available_packages()
    print(f"📦 Found {len(pkgs)} available language packages\n")

    for fr, to in LANG_PAIRS:
        print(f"=== {fr} → {to} ===")
        p = next((x for x in pkgs if x.from_code == fr and x.to_code == to), None)

        if not p:
            print(f"❌ Package NOT FOUND for {fr} → {to}\n")
            continue

        try:
            print(f"⬇ Downloading {fr}->{to}...")
            path = p.download()

            print("📥 Installing...")
            argostranslate.package.install_from_path(path)

            print(f"✅ Installed {fr} → {to}\n")
        except Exception as e:
            print(f"❌ Failed to install {fr}->{to}: {e}\n")

    print("\n🎉 All done!")

if __name__ == "__main__":
    main()
````

# Модуль обработки аудио

В данном модуле представлены следующие модели:
- **ASR - whisper medium**, в отличие от small показала полностью адекватное распознавание речи;

- **Translator - argostranslate**, легкие и быстрые модели для перевода локально

- **TTX - XTTS-v2** - мультиязычная модель с локальной обработкой аудио и возможностью сохранения характеристик голоса

````python
%%writefile tts_module.py
# ============================================
# OFFLINE ASR + OFFLINE TRANSLATION + XTTS-v2
# ============================================

import numpy as np
import librosa
import soundfile as sf
from scipy.signal import welch, lfilter
import tempfile
import os
os.environ["COQUI_TOS_AGREED"] = "1"    # автоматическое принятие соглашения для XTTS

# -----------------------------
# Оффлайн Whisper (openai-whisper)
# -----------------------------
import whisper

# -----------------------------
# Оффлайн перевод ArgosTranslate
# -----------------------------
import argostranslate.translate as argos_t


# -----------------------------
# Coqui XTTS-v2
# -----------------------------
from TTS.api import TTS


# ============ Голосовые утилиты ============
def estimate_formants(y, sr, n_formants=5):
    """Приближённая оценка формант через LPC."""
    try:
        n = int(0.03 * sr)
        frame = y[:n] if len(y) >= n else np.pad(y, (0, n-len(y)))
        frame = lfilter([1, -0.97], [1], frame)

        from librosa import lpc
        a = lpc(frame, int(2 + sr / 1000))
        roots = np.roots(a)
        roots = [r for r in roots if np.imag(r) > 0]
        angz = np.angle(roots)
        freqs = sorted(angz * (sr / (2*np.pi)))
        return freqs[:n_formants]
    except:
        return []


def compute_mfcc(y, sr):
    return librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)


def compute_spd(y, sr):
    return welch(y, sr, nperseg=1024)


def compute_zcr(y):
    return librosa.feature.zero_crossing_rate(y)[0]


# ============================================
#                MAIN CLASS
# ============================================

class RealTimeTranslator:
    """
    Полностью оффлайн:
    - Whisper ASR
    - ArgosTranslate перевод
    - XTTS-v2 синтез
    - Анализ речевых характеристик
    """

    def __init__(self, device="cuda"):
        self.device = device

        # Whisper — оффлайн
        print("Loading Whisper model (medium)...")
        self.whisper = whisper.load_model("medium", device=device)

        # XTTS-v2 — оффлайн
        print("Loading XTTS-v2...")
        self.tts = TTS(
            model_name="tts_models/multilingual/multi-dataset/xtts_v2",
            progress_bar=True,  # Добавляем прогресс-бар для отслеживания
            gpu=True if device == "cuda" else False
        )
        print("✅ XTTS-v2 loaded (license accepted)")

    # --------------------------
    # ASR
    # --------------------------
    def transcribe(self, y, sr, lang="en"):
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp:
            sf.write(tmp.name, y, sr)
            result = self.whisper.transcribe(tmp.name)
            return result["text"]

    # --------------------------
    # Offline ArgosTranslate
    # --------------------------
    def translate(self, text, src, tgt):
        try:
            translated = argos_t.translate(text, src, tgt)
            return translated
        except Exception as e:
            print("Translation error:", e)
            return "(translation unavailable)"

    # --------------------------
    # XTTS-v2
    # --------------------------
    def synthesize(self, text, speaker_wav, lang, out_path="tts_out.wav"):
        wav = self.tts.tts(
            text=text,
            speaker_wav=speaker_wav,
            language=lang
        )
        sf.write(out_path, wav, self.tts.synthesizer.output_sample_rate)
        return out_path

    # --------------------------
    # Анализ
    # --------------------------
    def analyze(self, original_path, synth_path):
        y1, sr1 = librosa.load(original_path, sr=None)
        y2, sr2 = librosa.load(synth_path, sr=None)

        # новое API librosa → resample по именам
        if sr1 != sr2:
            y2 = librosa.resample(y=y2, orig_sr=sr2, target_sr=sr1)
            sr2 = sr1

        return {
            "mfcc_o": compute_mfcc(y1, sr1),
            "mfcc_s": compute_mfcc(y2, sr2),
            "formants_o": estimate_formants(y1, sr1),
            "formants_s": estimate_formants(y2, sr2),
            "spd_o": compute_spd(y1, sr1),
            "spd_s": compute_spd(y2, sr2),
            "zcr_o": compute_zcr(y1),
            "zcr_s": compute_zcr(y2),
            "sr": sr1
        }
````

# Загрузка моделей

````python
%%writefile models.py
import torch
import gc

device_choice = "cuda" if torch.cuda.is_available() else "cpu"
_translator = None
_loaded = False

def get_translator():
    global _translator, _loaded

    if _translator is None:
        print(f"Загрузка моделей на {device_choice}...")
        from tts_module import RealTimeTranslator
        _translator = RealTimeTranslator(device=device_choice)
        _loaded = True
        print("Модели загружены!")

        gc.collect()
        if device_choice == "cuda":
            torch.cuda.empty_cache()

    return _translator

def load_models():
    """Функция для явной загрузки моделей"""
    global _loaded
    try:
        get_translator()
        _loaded = True
        return True
    except Exception as e:
        print(f"Ошибка загрузки моделей: {e}")
        return False

def are_models_loaded():
    """Проверка, загружены ли модели"""
    return _loaded
````

# Интерфейс Gradio

````python
%%writefile app.py
import gradio as gr
import numpy as np
import librosa
import matplotlib.pyplot as plt
import soundfile as sf
import io
from PIL import Image
import traceback
import torch
import gc
import os

# Добавляем согласие ПЕРЕД импортом TTS
os.environ["COQUI_TOS_AGREED"] = "1"

from models import get_translator, load_models, are_models_loaded

LANGS = {"EN": "en", "RU": "ru", "DE": "de", "FR": "fr"}

def load_models_ui():
    """Функция для загрузки моделей через UI"""
    try:
        if are_models_loaded():
            return "✅ Модели уже загружены!"

        success = load_models()
        if success:
            return "✅ Модели успешно загружены!"
        else:
            return "❌ Ошибка загрузки моделей"
    except Exception as e:
        return f"❌ Ошибка: {str(e)}"

def process_audio(audio_file, src_lang, tgt_lang, device):
    if audio_file is None:
        return "Загрузите аудиофайл", None, None, None, None, ""

    # Проверяем, загружены ли модели
    if not are_models_loaded():
        return "Сначала загрузите модели (кнопка 'Загрузить модели')", None, None, None, None, ""

    try:
        trans = get_translator()

        # Чтение входного аудио
        y, sr = sf.read(audio_file)
        if y.ndim > 1:
            y = np.mean(y, axis=1)

        sf.write("orig.wav", y, sr)

        # График waveform оригинала
        fig0 = plt.figure(figsize=(10,2))
        plt.plot(np.linspace(0, len(y)/sr, len(y)), y)
        plt.title("Оригинальное аудио (waveform)")
        buf0 = io.BytesIO()
        fig0.savefig(buf0, format='png')
        buf0.seek(0)
        img0 = Image.open(buf0)
        plt.close(fig0)

        # ASR → Translate → TTS
        full_asr = trans.transcribe(y, sr, LANGS[src_lang])
        translated = trans.translate(full_asr, LANGS[src_lang], LANGS[tgt_lang])
        out_path = trans.synthesize(translated, "orig.wav", LANGS[tgt_lang], "tts.wav")

        # Waveform синтезированного
        y_tts, sr_tts = librosa.load(out_path, sr=None)

        fig1 = plt.figure(figsize=(10,2))
        plt.plot(np.linspace(0, len(y_tts)/sr_tts, len(y_tts)), y_tts)
        plt.title("Синтезированный аудиосигнал (waveform)")
        buf1 = io.BytesIO()
        fig1.savefig(buf1, format='png')
        buf1.seek(0)
        img1 = Image.open(buf1)
        plt.close(fig1)

        # Анализ SPD
        analysis = trans.analyze("orig.wav", "tts.wav")
        fig2 = plt.figure(figsize=(10,3))
        freq_o, spd_o = analysis["spd_o"]
        freq_s, spd_s = analysis["spd_s"]
        plt.semilogy(freq_o, spd_o, label="original")
        plt.semilogy(freq_s, spd_s, label="synth")
        plt.legend()
        buf2 = io.BytesIO()
        fig2.savefig(buf2, format='png')
        buf2.seek(0)
        img2 = Image.open(buf2)
        plt.close(fig2)

        # Освобождение CUDA
        gc.collect()
        if device == "cuda":
            torch.cuda.empty_cache()

        return full_asr, translated, audio_file, out_path, [img0, img1], img2

    except Exception as e:
        err = f"Ошибка обработки: {str(e)}\n{traceback.format_exc()}"
        print(err)
        return err, None, None, None, None, ""

# ------------------------------
#      UI Gradio
# ------------------------------

with gr.Blocks() as demo:
    gr.Markdown("# 🎧 Real-Time Speech Translator")

    # Секция загрузки моделей
    gr.Markdown("### Шаг 1: Загрузка моделей")
    load_btn = gr.Button("📥 Загрузить модели", variant="primary")
    load_status = gr.Textbox(label="Статус загрузки", value="Модели не загружены")

    gr.Markdown("### Шаг 2: Обработка аудио")

    audio_input = gr.File(label="Загрузите аудио (wav/mp3/ogg/m4a)", type="filepath")

    with gr.Row():
        src_lang = gr.Dropdown(["EN", "RU", "DE", "FR"], label="Исходный язык", value="RU")
        tgt_lang = gr.Dropdown(["EN", "RU", "DE", "FR"], label="Целевой язык", value="EN")

    device = gr.Radio(["cuda", "cpu"], label="Устройство", value="cuda")

    process_btn = gr.Button("▶ Запустить обработку")

    asr_output = gr.Textbox(label="ASR")
    trans_output = gr.Textbox(label="Перевод")
    orig_audio = gr.Audio(label="Оригинал")
    tts_audio = gr.Audio(label="TTS")
    plots = gr.Gallery(label="Графики")
    analysis_output = gr.Image(label="SPD")

    # Обработчики событий
    load_btn.click(
        load_models_ui,
        outputs=load_status
    )

    process_btn.click(
        process_audio,
        inputs=[audio_input, src_lang, tgt_lang, device],
        outputs=[asr_output, trans_output, orig_audio, tts_audio, plots, analysis_output]
    )

demo.launch(
    share=True,
    debug=True,
    show_error=True,
    enable_monitoring=False,
    prevent_thread_lock=True,
)
````

На основе анализа графика SPD можно сделать следующие выводы:

2. Базовое соответствие:

    Модель правильно определяет основную спектральную массу (где сосредоточена энергия).

3. Отсутствие грубых искажений:

    Не видно сильных посторонних "выбросов" или провалов, которых нет в оригинале. Это говорит об отсутствии критических артефактов вроде сильного дребезжания или разрывов.



**Почему для XTTS-v2 это норма:**

1. **Архитектурная наследственность:**

    XTTS-v2 основана на VITS (а та, в свою очередь, на Glow-TTS и HiFi-GAN). Это потоковая (flow-based) модель, которая в своей основе оптимизирует функцию максимального правдоподобия. Такие модели печально известны склонностью к "оверсмутингу" (oversmoothing) — они выдают "усреднённый", наиболее вероятный спектр, теряя высокочастотные детали и шумовые компоненты, которые делают речь живой.

2. **Фокус на разборчивости и стабильности:**

    Основная цель XTTS (особенно в её версии 2) — предоставить стабильную, работающую из коробки, мультиязычную модель. Она жертвует детализацией спектра ради:

- Отсутствия "зацикливаний" и артефактов прерывания речи.

- Управляемости и консистентности.

- Возможности работать на относительно скромном железе (по сравнению с гигантскими диффузионными моделями).

- Быстрого инференса.

3. **Мультиязычность усугубляет компромисс:**

    Модель вынуждена находить общие паттерны для десятков языков с разной фонетикой и просодией. Это часто приводит к ещё большему усреднению и сглаживанию специфических для языка или диктора деталей, чтобы избежать катастрофического забывания или нестабильности на каком-то одном языке.

**Сравнение с другими моделями:**

**XTTS-v2 (VITS-like)** - хорошая формантная структура, но сглаженные детали и заваленные высокие частоты. Сила: стабильность, скорость, мультиязычность.

**Диффузионные модели (например, Parler-TTS, Stable Audio)** - гораздо более детализированный и "шумоподобный", естественный спектр, близкий к оригиналу. **Слабость** - очень медленный синтез, сложность управления, могут быть артефакты.

**Большие языковые модели для аудио (например, Voicebox, MMS)** - могут достигать лучшей детализации, но требуют колоссальных вычислительных ресурсов.

**Коммерческие кодеки (например, Descript, Adobe)** - часто используют гибридные подходы (VQVAE+GAN) и пост-обработку, чтобы добиться и детализации, и скорости.
