## Задача 1. Книги без буквы E.

**Условие.** В истории литературы известен случай написания романа объемом около
50 тыс. слов, в котором ни разу не была употреблена самая популярная
в английском алфавите буква E. Название его – «Gadsby».
Напишите программу, которая будет считывать список слов из файла и собирать статистику о том, в каком проценте слов используется каждая буква алфавита. Выведите результат для всех 26 букв английского алфавита и отдельно отметьте букву, которая встречалась в словах наиболее редко. В  вашей программе должны игнорироваться знаки препинания и регистр символов.


**Подсказки.**
1. Переведите все слова в верхний регистр:
```
word = word.upper().rstrip() # upper - переводит строку в верхний регистр
```
2. Создайте словарь со счетчиком слов, содержащий каждую букву
```
# Для каждой буквы инициализируем счетчик нулем
counts = {}
for ch in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
       counts[ch] = 0
```
3. Наиболее редко используемые буквы можно найти следующим образом:
```
smallest_count = min(counts.values()) # min - возвращает наименьшее число в списке
```

````python
#Напишите свое решение

# использовал ресурс https://www.gutenberg.org/cache/epub/47342/pg47342-images.html
# скачал текстовым файлом, и удалил весь вводный и закрывающий текст к книге, оставил текст самой истории



# создаем словарь, и заполняем его ключами - алфавитом, и значениями = 0
counts = {}
for ch in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
    counts[ch] = 0

# откроем файл в режиме чтения
with open("Gadsby.txt", "r") as file:
    lines = file.readlines()        # прочитаем весь файл построчно
    for line in lines:              # считаем каждую строку в цикле
        for ch in line.upper():     # в каждой строке с верхним регистром
            if ch in counts:        # читаем каджый символ
                counts[ch] += 1     # если он есть в словаре, добавляем 1 к его счетчику


smallest_count = min(counts.values())   # выбираем минимальное значение в словаре

# перебираем по ключам их значения, когда найдется минимальное
# то присвоим параметру key имя этого ключа в словаре
key = next(key for key, value in counts.items() if value == smallest_count)

# выведем наименее использующийся символ
print('наименее использованный символ:', key, smallest_count)

# выведем весь словарь построчно
for ch in counts:
    print(ch, counts[ch])
````

## Задание 2. Веб-скрейпинг ИИ новостей

**Описание задания.** Используя пример веб-скрейпинга практической части урока, проведите скрейпинг сайта с новостями из сферы ИИ (https://2051.vision/category/ii/), выведите на экран заголовки новостей.

````python
#Напишите свое решение
# импорт библиотек
from bs4 import BeautifulSoup
import requests

# Добавим заголовок, чтобы наш запрос распознавался как от браузера
headers = { "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:102.0) Gecko/20100101 Firefox/102.0" }

url = ("https://2051.vision/category/ii/") # адрес сайта
html = requests.get(url).text              # запрашиваем текстовый DOM
soup = BeautifulSoup(html, 'html5lib')     # Применяем к данным анализатор html5lib

# найдем заголовки 3 уровня с классом
top_news = soup.find_all('h3', class_='entry-title')

filteredNews=[]

# найдем все ссылки в заголовках и добавим их в лист
for data in top_news:
    if data.find('a') is not None:
        filteredNews.append(data.text.strip())

# выведем наш лист пронумерованным списком
for i, news in enumerate(filteredNews):
    print(i+1, news)

````
