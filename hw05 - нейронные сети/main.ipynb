## Задача 1. Класс Neuron

**Условие** Используя примеры из теоретической части 5.1 и 5.2, составьте класс Neuron модели нейрона, принимающего на вход от одного до 10 входных параметров. Запрещается пользоваться библиотекой numpy. На вход подается список. Матричное умножение реализуется с помощью циклов самостоятельно. Класс содержит методы для вычисления взвешенной суммы, а также содержит функции активации из урока.

```python
#Напишите свое решение
class Neuron:

    def __init__ (self, w):
        if 1<= len(w) <= 10:     # ограничиваем количество входных параметров
            self.w = w
        else:
            return 'количество входных параметров для класса Neuron от 1 до 10'

    def summ(self, x):
        s=0
        for i in range(len(x)):     # перемножаем векторы входов и веса
            s = s + self.w[i]*x[i]
        return s

    def onestep(self, x, bias):        # решаем, выполнять или нет, сравнив сумму весов
                                    # и величину смещения - наше пороговое значение,
        if int(x) >= bias:                  # выше которого надо передать сигнал дальше
            return 1, 'идем'
        else:
            return 0, 'не идем'
```

```python
# проверка цикла, на выходе лист, нужен int -> переделаем s=[] в s=0,
a=[1,2,1,3,4]
b=[2,2,2,2,2]
s=[]
for i in range(len(b)):
    s = s + [a[i]*b[i]]

print(s)
```
