##TF-IDF Vectorizer

TF-IDF (Term Frequency-Inverse Document Frequency) Vectorizer — это метод преобразования текстовых данных в числовые векторы, которые можно использовать в алгоритмах машинного обучения. TF-IDF учитывает не только частоту термина в документе, но и его важность в корпусе документов.

**Term Frequency** (TF): Частота термина в документе. Рассчитывается как количество раз, которое слово появляется в документе, деленное на общее количество слов в документе.

**Inverse Document Frequency** (IDF): Обратная частота документа. Рассчитывается как логарифм отношения общего количества документов к количеству документов, содержащих данный термин.

**TF-IDF**: Произведение TF и IDF.



###Применение

Для использования необходимо подключить

`from sklearn.feature_extraction.text import TfidfVectorizer`

Метод использования:

```
#Создание TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Преобразование текстов в TF-IDF матрицу
X = vectorizer.fit_transform(text)

# второй вариант преобразования
X = vectorizer.transform(text)

# Получение имен признаков
feature_names = vectorizer.get_feature_names_out()

# Визуализация
print("Имена признаков (терминов):")
print(feature_names)

# Создание DataFrame для визуализации
df = pd.DataFrame(X.toarray(), columns=feature_names)
print("\nМатрица признаков:")
print(df)
```

Есть 2 метода преобразования:
1.  **vectorizer.fit_transform(text)**
- **Fit**: Анализирует и обучает векторизатор на предоставленных данных. В случае TfidfVectorizer это означает, что он определяет словарь всех уникальных слов и вычисляет статистику для TF-IDF на основе обучающих данных.

- **Transform**: Преобразует текстовые данные в матрицу TF-IDF на основе обученного словаря и статистики.

2. **vectorizer.transform(text)** - Этот метод только преобразует текстовые данные в матрицу TF-IDF, используя уже обученный словарь и статистику. Он не изменяет и не переобучает векторизатор.

###Параметры TF-IDF Vectorizer:

**max_df**: Максимальная частота документа. Термины, встречающиеся чаще этого порога, игнорируются. (Полезно для отсеивания слов, которые встречаются почти везде, но не несут полезной нагрузки для классификации)

**min_df**: Минимальная частота документа. Термины, встречающиеся реже этого порога, игнорируются.

**max_features**: Максимальное количество признаков (терминов) для включения.

**stop_words**: Список стоп-слов для исключения. (может принимать как свой словарь слов, так и язык документа)

**ngram_range**: Диапазон n-грамм для включения (например, (1, 2) для униграмм и биграмм).

**use_idf**: Включение или отключение IDF.

**smooth_idf**: Добавление сглаживания к IDF.

**sublinear_tf**: Применение сублинейного масштабирования к TF.

##PassiveAggressiveClassifier

**PassiveAggressiveClassifier** — это алгоритм линейной классификации, который предназначен для больших и потоковых данных. Он работает на основе онлайн-обучения, обновляя модель при каждом новом примере.

###Особенности PassiveAggressiveClassifier:

**Passive**: Если предсказание правильное, модель остается неизменной.

**Aggressive**: Если предсказание неправильное, модель обновляется, чтобы исправить ошибку.
Подходит для задач классификации и регрессии.

Быстро обновляется при каждом новом примере, что делает его эффективным для потоковых данных.

###Применение

```
# Создание и обучение модели PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=1000, random_state=42)
pac.fit(X_train, y_train)

# Предсказание на тестовых данных
y_pred = pac.predict(X_test)

# Оценка точности модели
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

###Параметры

**C**: Параметр регуляризации. Чем меньше значение, тем сильнее регуляризация.

**fit_intercept**: Включение или отключение константного члена (свободного члена).

**max_iter**: Максимальное количество итераций для обучения.

**tol**: Допуск для остановки обучения.

**early_stopping**: Остановка обучения при отсутствии улучшения.

**validation_fraction**: Доля данных для валидации при использовании ранней остановки.

**n_iter_no_change**: Количество итераций без улучшения для остановки обучения.

**shuffle**: Перемешивание данных перед каждой эпохой.

###Предобработка данных

Уберем все знаки препинания в тексте, приведем его в нижний регистр и прогоним наш текст через TF-IDF vectorizer, чтобы получить векторы слов.

````python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

import re
import string
````

````python
# функция очистки текста
def preprocess(text):
    # удаление знаков препинания и приведение в нижний регистр
    text = re.sub(f'[{string.punctuation}]', '', text.lower())
    return text
````

Пояснение:

**string.punctuation** содержит все знаки пунктуации, определенные в модуле string.

**f'[{string.punctuation}]'** создает строку формата (f-string), которая содержит все знаки пунктуации в квадратных скобках, что делает её регулярным выражением для поиска любых знаков пунктуации.

**re.sub(pattern, repl, string)** заменяет все вхождения шаблона **pattern** в строке **string** на строку **repl**. В данном случае шаблоном является **f'[{string.punctuation}]'**, а заменой является пустая строка ' '.

````python
# применяем функцию к тексту
df['text'] = df['text'].apply(preprocess)
````

````python
# разделим текст на выборки для обучения и теста
x_train, x_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2)
````

````python
# применяем TF-IDF V
tfidf_v = TfidfVectorizer(stop_words='english', max_df=0.7)

tfidf_train = tfidf_v.fit_transform(x_train)
tfidf_test = tfidf_v.transform(x_test)
````

###Обучение модели Passive Agressive Classifier

Для обучения модели мы будет использовать несколько вариантов настроек, и выберем наилучшие, основываясь на показателях точности.

Для данной цели нам подойдет **GridSearchCV** с параметром кросс-валидации (перекрестной проверки) `cv=StratifiedKFold`

- **Параметр CV** определяет количество фолдов (подмножеств), на которые будет разделен набор данных. Например, если cv=5, то данные будут разделены на 5 фолдов. В каждом из 5 экспериментов модель будет обучаться на 4 фолдах и тестироваться на оставшемся фолде. Этот процесс повторяется 5 раз, и результаты усредняются для получения оценки производительности модели.

    **Целое число** (например, cv=5): Данные будут разделены на указанное количество фолдов. Это наиболее часто используемый вариант.

    **Объект перекрестной проверки** (например, StratifiedKFold): Вы можете передать объект, который реализует стратегию перекрестной проверки. Это полезно, если вы хотите использовать специфические методы перекрестной проверки, такие как стратифицированная перекрестная проверка (StratifiedKFold), которая сохраняет пропорции классов в каждом фолде.

    **Другие стратегии:** Можно использовать более сложные стратегии, такие как LeaveOneOut, LeavePOut, и т.д.

- **Почему StratifiedKFold?**
    
    **Сбалансированное распределение классов:** StratifiedKFold сохраняет пропорции классов в каждом фолде, что особенно важно для задач классификации, где классы могут быть несбалансированными.

    **Надежная оценка:** Использование 5 или 10 фолдов обычно обеспечивает надежную оценку производительности модели, не увеличивая при этом слишком сильно время обучения.

````python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
````

````python
pac = PassiveAggressiveClassifier()

# определение набора значений для max_iter
param_grid = {'max_iter' : [10, 20, 50, 100, 200, 500, 1000]}

# задание объекта перекрестной проверки
skf = StratifiedKFold(n_splits=5)

# использование GridSearchCV для задания подбора гиперпараметров
grid_search = GridSearchCV(pac, param_grid, cv=skf)
grid_search.fit(tfidf_train, y_train)

# выбор лучшей модели и итераций
best_model = grid_search.best_estimator_
best_max_iter = grid_search.best_params_['max_iter']

print(f'лучшее значение max_iter: {best_max_iter}')

# проверка точности лучшей модели
test_score = best_model.score(tfidf_test, y_test)
print(f'точность на тестовой выборке: {test_score*100:.2f}%')
````

При нескольких выполнениях кода алгоритм выдает разные варианты лучшего значения, от 10 до 1000, но точность остается на уровне 95.5 - 95.9%

Попробуем провести тестовые обучения на разных max_iter сами и построим графики оценки точности от этого параметра

````python
import matplotlib.pyplot as plt

max_iter_values = [10, 20, 50, 100, 200, 500, 1000]
acc_scores=[]

for max_iter in max_iter_values:
    pac=PassiveAggressiveClassifier(max_iter=max_iter)
    pac.fit(tfidf_train, y_train)
    y_pred=pac.predict(tfidf_test)
    acc_scores.append(accuracy_score(y_test, y_pred))

# строим графики
plt.plot(max_iter_values, acc_scores, marker='o')
plt.xlabel('количество итераций')
plt.ylabel('точность')
plt.title('график точности от итераций')
plt.show()
````

Несколько построений графика показывают, что точность сохраняется в пределах 95,5 - 96%, кроме случаев, когда алгоритм не успевает сойтись.

##XGBoost

**XGBoost (eXtreme Gradient Boosting)** — это мощный и эффективный инструмент для машинного обучения, который используется для задач классификации и регрессии. Он основан на технике градиентного бустинга, которая строит ансамбль слабых моделей (обычно деревьев решений) последовательно, каждая из которых исправляет ошибки предыдущих моделей.

###Особенности

1. **Эффективность и производительность:**

    - **Высокая производительность:** XGBoost оптимизирован для параллельного и распределенного вычисления, что значительно ускоряет обучение моделей.
    - **Сжатие памяти:** Использует оптимизированные структуры данных, такие как DMatrix, для уменьшения потребления памяти и увеличения скорости.

2. **Обработка пропущенных значений:**

    XGBoost автоматически обрабатывает пропущенные значения в данных, что упрощает предварительную обработку данных.
3. **Регуляризация:**

    Включает L1 (Lasso) и L2 (Ridge) регуляризацию, что помогает предотвратить переобучение моделей.

4. **Поддержка различных функций потерь:**

    Поддерживает множество функций потерь, включая логистическую регрессию для классификации и среднеквадратичную ошибку для регрессии.
5. **Раннее завершение (Early Stopping):**

    Позволяет остановить обучение, если модель перестает улучшаться на валидационном наборе данных, что помогает избежать переобучения.
6. **Поддержка кастомных функций потерь и оценок:**

    Пользователи могут задавать свои собственные функции потерь и оценки, что делает XGBoost очень гибким.

###Области использования
XGBoost используется в различных областях, включая:

- Финансовый анализ: кредитный скоринг, прогнозирование цен на акции.

- Медицинская диагностика: предсказание исходов заболеваний, анализ медицинских данных.

- Маркетинг и реклама: сегментация клиентов, предсказание отклика на рекламные кампании.

- Интернет-технологии: рекомендации, анализ поведения пользователей.

- Соревнования по машинному обучению: XGBoost часто показывает отличные результаты на конкурсах, таких как Kaggle.

###Основные параметры XGBoost

**Общие параметры:**

**`booster`:** Тип бустера для использования (gblin, gbtree, dart).

**`nthread`:** Количество потоков для использования.

**Параметры дерева решений:**

**`max_depth`:** Максимальная глубина дерева (по умолчанию 6).

**`min_child_weight`:** Минимальный вес для дочернего узла (по умолчанию 1).

**`gamma`:** Минимальное уменьшение потерь, необходимое для разделения узла (по умолчанию 0).

**`subsample`:** Доля выборки для каждого дерева (по умолчанию 1).

**`colsample_bytree`:** Доля фич для каждого дерева (по умолчанию 1).

**Параметры линейного бустера:**

**`lambda`:** L2 регуляризация (по умолчанию 1).

**`alpha`:** L1 регуляризация (по умолчанию 0).

**Параметры обучения:**

**`eta`**: Скорость обучения (по умолчанию 0.3).

**`objective`:** Функция потерь (например, reg:squarederror для регрессии, binary:logistic для бинарной классификации).

**`eval_metric`:** Метрика для оценки модели (например, rmse, logloss).

**Параметры раннего завершения:**

**`early_stopping_rounds`:** Количество раундов без улучшений для остановки обучения.

###Модели XGBoost
Модели XGBoost могут быть адаптированы для различных задач, таких как регрессия, классификация, ранжирование и многоклассовая классификация. В зависимости от конкретной задачи, следует выбирать соответствующий тип модели и настраивать параметры. Вот основные модели XGBoost и их применение:

1. **Регрессия (Regression)**

    Для задач регрессии, где нужно предсказать непрерывное значение, используется XGBRegressor.

    **`model = xgb.XGBRegressor(objective='reg:squarederror')`**

2. **Бинарная классификация (Binary Classification)**

    Для задач бинарной классификации, где нужно предсказать одно из двух значений (например, да/нет, 0/1), используется XGBClassifier.

    **`model = xgb.XGBClassifier(objective='binary:logistic')`**

3. **Многоклассовая классификация (Multiclass Classification)**

    Для задач многоклассовой классификации, где нужно предсказать одно из нескольких значений (например, категории), также используется XGBClassifier, но с параметром **objective='multi:softmax'** или **objective='multi:softprob'**.

    **multi:softmax** возвращает метку класса с наибольшей вероятностью.

    **multi:softprob** возвращает вероятности для всех классов.

    **`model = xgb.XGBClassifier(objective='multi:softmax', num_class=3)`**

4. **Ранжирование (Ranking)**

    Для задач ранжирования, таких как ранжирование документов в поисковой системе, используется XGBRanker.

    **`model = xgb.XGBRanker(objective='rank:pairwise')`**

##Метрики и их значение

Для определения качества обучения модели применяются различные метрики, рассмотрим наиболее часто встречающиеся:

**Бинарная классификация**

- **Accuracy (Точность):** Процент правильно классифицированных примеров.

- **Precision (Точность):** Доля истинно положительных результатов среди всех примеров, классифицированных как положительные.

- **Recall (Полнота):** Доля истинно положительных результатов среди всех положительных примеров.

- **F1-Score:** Гармоническое среднее между точностью и полнотой.

- **ROC-AUC (Area Under the Receiver Operating Characteristic Curve):** Площадь под кривой ROC, показывающая способность модели различать положительные и отрицательные классы.

**Мультиклассовая классификация**

- **Accuracy (Точность):** Процент правильно классифицированных примеров.

- **Precision (Точность), Recall (Полнота) и F1-Score:** Эти метрики могут быть вычислены для каждого класса и усреднены (например, среднее по макро или микро).

- **Logloss (Logarithmic Loss):** Метрика, которая измеряет неопределённость модели в предсказаниях. Чем ниже значение, тем лучше.

###Сравнение и применение

**Accuracy (Точность)** измеряет долю правильно классифицированных примеров среди всех примеров.

**Преимущества:**

- Простая и интуитивно понятная метрика.
Хорошо работает, когда классы сбалансированы, то есть количество примеров каждого класса примерно одинаково.

**Недостатки:**

- Может вводить в заблуждение, если классы несбалансированы. Например, если один класс сильно преобладает, модель может показывать высокую точность, просто предсказывая этот класс для всех примеров.
---

**Precision (точность предсказаний)** измеряет долю истинно положительных результатов среди всех примеров, классифицированных как положительные.

**Преимущества:**

- Полезна, когда важно минимизировать количество ложноположительных результатов (False Positives).
- Часто используется в задачах, где ложноположительные результаты могут иметь высокую стоимость (например, в медицинской диагностике или спам-фильтрах).

**Недостатки:**

- Не учитывает ложные отрицательные результаты (False Negatives), что может быть важно в некоторых задачах.
---
**Recall (полнота)** измеряет долю истинно положительных результатов среди всех реальных положительных примеров.

**Преимущества:**

- Полезна, когда важно минимизировать количество ложноотрицательных результатов (False Negatives).
- Часто используется в задачах, где пропуск положительных примеров может иметь высокую стоимость (например, в медицинской диагностике, где важно не пропустить болезнь).

**Недостатки:**

- Не учитывает ложноположительные результаты (False Positives), что может быть важно в некоторых задачах.
---
**F1-Score** является гармоническим средним между Precision и Recall. Эта метрика пытается сбалансировать Precision и Recall, предоставляя единое значение, которое учитывает как ложноположительные, так и ложноотрицательные результаты.

**Преимущества:**

- Полезна, когда нужно учитывать как Precision, так и Recall, особенно в случаях, когда имеется дисбаланс между классами.
- Предоставляет единое значение для оценки модели, что может быть удобным для сравнения моделей.

**Недостатки:**

- Может быть сложнее интерпретировать по сравнению с отдельными метриками Precision и Recall.
---

**Пример**

Рассмотрим пример, чтобы лучше понять разницу между этими метриками.

Предположим, у нас есть модель для классификации спама в электронной почте. У нас есть следующие результаты:

True Positives (**TP**): 70 (спам правильно классифицирован как спам)

False Positives (**FP**): 10 (не спам ошибочно классифицирован как спам)

True Negatives (**TN**): 900 (не спам правильно классифицирован как не спам)

False Negatives (**FN**): 20 (спам ошибочно классифицирован как не спам)







$$
Accuracy = \frac{TP+TN}{TP+FP+TN+FN} = \frac{70+900}{70+10+900+20} = \frac{970}{1000} = 0.97
$$

$$
Precision = \frac{TP}{TP+FP} = \frac{70}{70+10} = \frac{70}{80} = 0.875
$$

$$
Recall = \frac{TP}{TP+FN} = \frac{70}{70+20} = \frac{70}{90} = 0.778
$$

$$
F1-Score =2 * \frac{Precision}{Recall} = 2 * \frac{0.875*0.778}{0.875+0.778} = \frac{0.681}{1.653} = 0.823
$$


**Accuracy (Точность):** 97% показывает, что модель в целом хорошо справляется с классификацией писем, но может быть менее полезна при дисбалансе классов.

**Precision (Точность предсказаний):** 87.5% указывает, что большинство писем, классифицированных как спам, действительно являются спамом, что важно для минимизации ложноположительных результатов.

**Recall (Полнота):** 77.8% указывает, что модель находит большинство спам-писем, что важно для минимизации ложноотрицательных результатов.

**F1-Score:** 82.4% балансирует между precision и recall, предоставляя общее представление о производительности модели.

###Заключение по метрикам

**Accuracy** измеряет общую долю правильных предсказаний и хорошо работает при сбалансированных классах.

**Precision** фокусируется на качестве положительных предсказаний и важна, когда нужно минимизировать ложноположительные результаты.

**Recall** фокусируется на способности модели находить все положительные примеры и важна, когда нужно минимизировать ложноотрицательные результаты.

**F1-Score** является гармоническим средним **Precision** и **Recall** и полезен, когда нужно сбалансировать эти две метрики, особенно в случае дисбаланса классов.

###Настройка параметров обучения

Данные во фрейме представлены не равноверно, по этому при обучении модели снова воспользуемся **StratifiedKFolds**.

В этот раз воспользуемся поиском лучших параметров с помощью **RandomizedSearchCV**. В отличие от GridSearchCV, который перебирает все заданные варианты настройки, RandomizedSearchCV выбирает комбинации параметров рандомно из набора представленных ему вариантов. Это ускоряет поиск при наличии большого числа настраиваемых параметров.

````python
# параметры для StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=12)

# параметры для RandomizedSearchCV
params = {
    'base_score': [0.1, 0.2, 0.5, 0.7],
    'n_estimators': [100, 200, 300, 400, 500],
    'learning_rate': [0.01, 0.05, 0.1, 0.2, 0.3, 0.4],
    'max_depth': [3, 4, 5, 6, 7, 8, 9],
    'subsample': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1],
    'colsample_bytree': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1],
    'importance_type': ['weight', 'gain', 'cover'],
}
````

Параметры **RandomizedSearchCV**:

- **`n_estimators`** - количество деревьев в модели. Чем больше деревьев, тем сложнее и потенциально более точной может стать модель, но это также увеличивает время обучения и риск переобучения.

- **`learning_rate`** - Скорость обучения модели (шаг градиента). Меньшие значения делают обучение более медленным, но могут привести к более точной модели. Большие значения ускоряют обучение, но могут привести к пропуску оптимального решения.

- **`max_depth`** - Максимальная глубина каждого дерева. Большая глубина позволяет модели захватывать более сложные зависимости в данных, но также увеличивает риск переобучения.

- **`subsample`** - Доля выборки, используемая для обучения каждого дерева. Меньшие значения могут помочь предотвратить переобучение, увеличивая разнообразие деревьев, но также могут снизить точность модели.

- **`colsample_bytree`** - Доля признаков, используемых для обучения каждого дерева. Меньшие значения могут помочь предотвратить переобучение, увеличивая разнообразие деревьев, но также могут снизить точность модели.

````python
# инициализация модели XGBoost
xgb = XGBClassifier(objective='binary:logistic',  eval_metric='logloss')

# инициализация RandomizedSearchCV
rand_search = RandomizedSearchCV(xgb, param_distributions=params, n_iter=100, scoring='accuracy', n_jobs=-1, cv=skf, random_state=10, verbose=1)
````

Дполонительные параметры:

- **`n_jobs`** - указывает количество параллельных процессов для обучения.

- **`verbose`** - включает визуазизацию обучения.
